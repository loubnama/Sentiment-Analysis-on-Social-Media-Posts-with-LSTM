```python
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
import numpy as np

!wget http://nlp.stanford.edu/data/glove.6B.zip
!unzip glove.6B.zip

def load_glove_embeddings(glove_file):
    embeddings_index = {}
    with open(glove_file, encoding='utf8') as f:
        for line in f:
            values = line.split()
            word = values[0]
            coefs = np.asarray(values[1:], dtype='float32')
            embeddings_index[word] = coefs
    return embeddings_index

embeddings_index = load_glove_embeddings('glove.6B.100d.txt')

tokenizer = Tokenizer()
tokenizer.fit_on_texts(X_train)
X_train_sequences = tokenizer.texts_to_sequences(X_train)
X_test_sequences = tokenizer.texts_to_sequences(X_test)
max_length = 100
X_train_padded = pad_sequences(X_train_sequences, maxlen=max_length)
X_test_padded = pad_sequences(X_test_sequences, maxlen=max_length)
vocab_size = len(tokenizer.word_index) + 1
embedding_dim = 100
embedding_matrix = np.zeros((vocab_size, embedding_dim))
for word, i in tokenizer.word_index.items():
    embedding_vector = embeddings_index.get(word)
    if embedding_vector is not None:
        embedding_matrix[i] = embedding_vector

print(f"Training set size: {len(X_train)}")
print(f"Testing set size: {len(X_test)}")
print(f"Shape of X_train_padded: {X_train_padded.shape}")
print(f"Shape of X_test_padded: {X_test_padded.shape}")
print(f"Embedding matrix shape: {embedding_matrix.shape}")

```

```python
#3
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, Dropout
from tensorflow.keras.optimizers import Adam

def build_model(embedding_dim, vocab_size, max_length, num_classes, lstm_units=128, dropout_rate=0.5, learning_rate=0.001):
    model = Sequential()
    model.add(Embedding(input_dim=vocab_size, output_dim=embedding_dim, input_length=max_length))
    model.add(LSTM(lstm_units, return_sequences=False, dropout=dropout_rate, recurrent_dropout=dropout_rate))
    model.add(Dense(num_classes, activation='softmax'))

    optimizer = Adam(learning_rate=learning_rate)
    model.compile(loss='sparse_categorical_crossentropy', optimizer=optimizer, metrics=['accuracy'])
    return model

embedding_dim = 100
vocab_size = len(tokenizer.word_index) + 1
max_length = 100
num_classes = 6

model = build_model(embedding_dim, vocab_size, max_length, num_classes)

model.summary()

```

```python
history = model.fit(X_train_padded, y_train, epochs=50, batch_size=64, validation_split=0.2)
```

```python
#4
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

test_loss, test_accuracy = model.evaluate(X_test_padded, y_test)
print(f"Test Accuracy: {test_accuracy}")

y_pred = model.predict(X_test_padded)
y_pred_classes = y_pred.argmax(axis=1)

unique_classes = sorted(y_test.unique())
print(f"Unique classes in y_test: {unique_classes}")

print(classification_report(y_test, y_pred_classes, labels=unique_classes, target_names=[f'Class {i}' for i in unique_classes]))

conf_matrix = confusion_matrix(y_test, y_pred_classes, labels=unique_classes)
plt.figure(figsize=(10, 7))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=[f'Class {i}' for i in unique_classes], yticklabels=[f'Class {i}' for i in unique_classes])
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```

```python
def build_advanced_model(embedding_dim, vocab_size, max_length, num_classes, lstm_units=128, dropout_rate=0.5, learning_rate=0.001):
    model = Sequential()
    model.add(Embedding(input_dim=vocab_size, output_dim=embedding_dim, input_length=max_length))
    model.add(LSTM(lstm_units, return_sequences=True, dropout=dropout_rate, recurrent_dropout=dropout_rate))
    model.add(LSTM(lstm_units, return_sequences=False, dropout=dropout_rate, recurrent_dropout=dropout_rate))
    model.add(Dense(num_classes, activation='softmax'))

    optimizer = Adam(learning_rate=learning_rate)
    model.compile(loss='sparse_categorical_crossentropy', optimizer=optimizer, metrics=['accuracy'])
    return model

model_advanced = build_advanced_model(embedding_dim, vocab_size, max_length, num_classes)

model_advanced.summary()

history_advanced = model_advanced.fit(X_train_padded, y_train, epochs=60, batch_size=64, validation_split=0.2)
```

```python
#5
from tensorflow.keras.callbacks import EarlyStopping

def build_and_train_model(embedding_dim, vocab_size, max_length, num_classes, lstm_units, dropout_rate, learning_rate, batch_size, epochs):
    model = Sequential()
    model.add(Embedding(input_dim=vocab_size, output_dim=embedding_dim, input_length=max_length))
    model.add(LSTM(lstm_units, return_sequences=False, dropout=dropout_rate, recurrent_dropout=dropout_rate))
    model.add(Dense(num_classes, activation='softmax'))

    optimizer = Adam(learning_rate=learning_rate)
    model.compile(loss='sparse_categorical_crossentropy', optimizer=optimizer, metrics=['accuracy'])

    early_stopping = EarlyStopping(monitor='val_loss', patience=2)

    history = model.fit(X_train_padded, y_train, epochs=epochs, batch_size=batch_size, validation_split=0.2, callbacks=[early_stopping])

    return model, history

configurations = [
    {'lstm_units': 64, 'dropout_rate': 0.3, 'learning_rate': 0.001, 'batch_size': 64, 'epochs': 1},
    {'lstm_units': 128, 'dropout_rate': 0.5, 'learning_rate': 0.001, 'batch_size': 64, 'epochs': 1},
    {'lstm_units': 128, 'dropout_rate': 0.5, 'learning_rate': 0.0005, 'batch_size': 128, 'epochs': 1},
    {'lstm_units': 256, 'dropout_rate': 0.5, 'learning_rate': 0.001, 'batch_size': 64, 'epochs': 1},
]

results = []

for config in configurations:
    print(f"Testing configuration: {config}")
    model, history = build_and_train_model(
        embedding_dim=100,
        vocab_size=len(tokenizer.word_index) + 1,
        max_length=100,
        num_classes=6,
        lstm_units=config['lstm_units'],
        dropout_rate=config['dropout_rate'],
        learning_rate=config['learning_rate'],
        batch_size=config['batch_size'],
        epochs=config['epochs']
    )

    test_loss, test_accuracy = model.evaluate(X_test_padded, y_test)
    print(f"Test Accuracy: {test_accuracy}")

    results.append({
        'configuration': config,
        'test_accuracy': test_accuracy
    })
```

```python
import pandas as pd

results_df = pd.DataFrame(results)
print(results_df)

import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))
plt.bar(range(len(results_df)), results_df['test_accuracy'], tick_label=[str(c) for c in results_df['configuration']])
plt.xlabel('Configuration')
plt.ylabel('Test Accuracy')
plt.title('Model Performance Across Different Configurations')
plt.xticks(rotation=45, ha='right')
plt.show()
```
